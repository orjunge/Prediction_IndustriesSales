### 매출데이터

```python
# 매출 데이터 컬럼 확인 ==> 모든 연도가 동일 컬럼임을 확인함
print("▶▶▶▶2022 매출 컬럼\n", sales_2022.columns)
print("▶▶▶▶2021 매출 컬럼\n", sales_2021.columns)
print("▶▶▶▶2020 매출 컬럼\n", sales_2020.columns)
print("▶▶▶▶2019 매출 컬럼\n", sales_2019.columns)
print("▶▶▶▶2018 매출 컬럼\n", sales_2018.columns)
print("▶▶▶▶2017 매출 컬럼\n", sales_2017.columns)
```

```python
### 매출 데이터 6개년 합치기 --> 열은 모두 똑같길래 concat 사용
dfs = [sales_2022, sales_2021, sales_2020, sales_2019, sales_2018, sales_2017]
sales_6years = pd.concat(dfs, ignore_index=True)


# 데이터 확인용 --> ex. '상권_코드'가 1001496일 경우 뭐가 나오나
check_test = sales_6years[sales_6years['상권_코드'] == 1001496]
check_test.head(30)
```

```python
### 상권코드에 맞춰 자치구, 행정동 끌어오기

# '자치구', '행정동' 열 생성
sales_6years['자치구'] = ''
sales_6years['행정동'] = ''

# 자치구, 행정동 채우기
for index, row in sales_6years.iterrows():
    mask = seoul_gudong['상권_코드'] == row['상권_코드']
    matching_data = seoul_gudong[mask]
    if not matching_data.empty:
        sales_6years.at[index, '자치구'] = matching_data['자치구'].values[0]
        sales_6years.at[index, '행정동'] = matching_data['행정동'].values[0]

sales_6years
```

### 매출+점포 병합

```python
# 두 데이터프레임 (sales, store)의 지정 컬럼의 값이 같을 때 점포데이터의 남은 컬럼 값 가져오기
df_merged = pd.merge(sales_6years, store_6years, on=['기준_년_코드', '기준_분기_코드', '상권_구분_코드', '상권_구분_코드_명', '상권_코드', '상권_코드_명', '서비스_업종_코드', '서비스_업종_코드_명'], how='inner')
```

```python
# 총 매출 금액 컬럼 생성
df_merged['총_매출_금액'] = df_merged['주중_매출_금액'] + df_merged['주말_매출_금액']
df_merged.columns
```

```python
# 필요한 컬럼만 추출
df_preprocessed = df_merged[['기준_년_코드', '기준_분기_코드', '자치구', '행정동', '상권_코드', '서비스_업종_코드', '서비스_업종_코드_명', '점포_수', '유사_업종_점포_수',
                             '개업_율', '개업_점포_수', '폐업_률', '폐업_점포_수', '프랜차이즈_점포_수',
                             '총_매출_금액', '분기당_매출_금액', '주중_매출_금액', '주말_매출_금액', '남성_매출_금액', '여성_매출_금액']]
df_preprocessed
```

### 대분류, 중분류 붙이기

```python
### '서비스_업종_코드_명'(업종명)에 따라 대분류, 중분류 채우기

# '대분류', '중분류' 컬럼 생성
df_preprocessed['대분류'] = ''
df_preprocessed['중분류'] = ''

# 중분류, 대분류 채우기
for index, row in df_preprocessed.iterrows():
    mask = category['업종명'] == row['서비스_업종_코드_명']
    matching_data = category[mask]
    if not matching_data.empty:
        df_preprocessed.at[index, '대분류'] = matching_data['대분류'].values[0]
        df_preprocessed.at[index, '중분류'] = matching_data['중분류'].values[0]

# 결과 확인
df_preprocessed
```

```python
# 필요한 컬럼만 추출
df_preprocessed = df_preprocessed[['기준_년_코드', '기준_분기_코드', '자치구', '행정동', '대분류', '중분류', '상권_코드', '서비스_업종_코드', '서비스_업종_코드_명', '점포_수', '유사_업종_점포_수',
                             '개업_율', '개업_점포_수', '폐업_률', '폐업_점포_수', '프랜차이즈_점포_수',
                             '총_매출_금액', '분기당_매출_금액', '주중_매출_금액', '주말_매출_금액', '남성_매출_금액', '여성_매출_금액']]
df_preprocessed.head()
```

```python
# 총 점포수 컬럼 추가
df_preprocessed['순점포증감률'] = (df_preprocessed['개업_점포_수'] - df_preprocessed['폐업_점포_수']) / df_preprocessed['유사_업종_점포_수']

# 컬럼 명 변경
df_preprocessed.rename(columns={'유사_업종_점포_수': '총_점포수'}, inplace=True)
df_preprocessed = df_preprocessed[['기준_년_코드', '기준_분기_코드', '자치구', '행정동', '대분류', '중분류', '상권_코드', '서비스_업종_코드', '서비스_업종_코드_명', '점포_수', '총_점포수', '순점포증감률', '개업_율', '개업_점포_수', '폐업_률', '폐업_점포_수',
                                   '프랜차이즈_점포_수', '총_매출_금액', '분기당_매출_금액', '주중_매출_금액', '주말_매출_금액', '남성_매출_금액', '여성_매출_금액']]
df_preprocessed.head(2)
```

```python
# '순점포증감률' 열의 NaN 값을 0으로 변경
df_preprocessed['순점포증감률'] = df_preprocessed['순점포증감률'].fillna(0)
df_preprocessed[df_preprocessed['순점포증감률'].isna()]
df_preprocessed.isnull().sum()
```
